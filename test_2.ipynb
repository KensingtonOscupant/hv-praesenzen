```python
import os
import pdfplumber
import pandas as pd
from tqdm import tqdm

def process_pdf(file_path):
    try:
        with pdfplumber.open(file_path) as pdf:
            full_text = ""
            for page in pdf.pages:
                page_text = page.extract_text()
                if page_text is None or page_text == "":
                    return None, "Document could not be read"
                full_text += page_text
            return full_text, None
    except Exception as e:
        return None, str(e)

def process_all_pdfs(directory):
    results = []
    pdf_files = [os.path.join(root, file)
                 for root, dirs, files in os.walk(directory)
                 for file in files if file.lower().endswith('.pdf')]

    for file_path in tqdm(pdf_files, desc="Processing PDFs"):
        text, error = process_pdf(file_path)
        if text is not None:
            results.append({'File Path': file_path, 'Character Count': len(text)})
        else:
            results.append({'File Path': file_path, 'Error': error})

    return pd.DataFrame(results)

# Replace 'your_directory_path' with the path to the directory you want to process
directory_path = 'data/Praesenzen_hv-info/'
df = process_all_pdfs(directory_path)

# Save the dataframe to a CSV file
csv_file_path = 'pdf_character_counts.csv'  # Replace with your desired file path
df.to_csv(csv_file_path, index=False)

print(f"Results saved to {csv_file_path}")
```

```python
import pandas as pd

# load data/Combined_new.csv
original_df = pd.read_csv('data/231221_Combined_new_with_ID.csv')
original_df

# load data/231223_hv_praesenzen_full_dataset.csv
results_df = pd.read_csv('data/231223_hv_praesenzen_full_gpt_response.csv')
results_df

# join the two dataframes on the ID column
df = pd.merge(original_df, results_df, on='unique_id', how='left')

# drop ID_Key_y and Year_y, header_row and all_percentage_values columns
df = df.drop(columns=['ID_Key_y', 'Year_y', 'header_row', 'all_percentage_values'])
df

# switch the location of of the unique_id and ID_Key_x columns
df = df[['unique_id', 'ID_Key_x', 'Company', 'Year_x', 'Presence_x', 'Index', 'Presence_y', 'error', 'cost', 'file_path', 'comment']]

# rename ID_Key_x to ID_Key and Year_x to Year and Presence_x to Presence_exisiting and Presence_y to Presence_GPT
df = df.rename(columns={'ID_Key_x': 'ID_Key', 'Year_x': 'Year', 'Presence_x': 'Presence_existing', 'Presence_y': 'Presence_GPT'})

df

# save to data/231223_hv_praesenzen_full_dataset.csv
df.to_csv('data/231223_hv_praesenzen_full_dataset.csv', index=False)
```
