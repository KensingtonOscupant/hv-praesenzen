```python
import pandas as pd

# Load the CSV file
file_path = 'data/231215_test_set_large.csv'  # Replace with your CSV file path
df = pd.read_csv(file_path)

# Group by 'ID_Key_original' and randomly pick one row from each group
random_selection = df.groupby('ID_Key_original').apply(lambda x: x.sample(1)).reset_index(drop=True)

# Save the new DataFrame to a CSV file
output_file_path = 'data/231219_test_set.csv'  # You can change the output file name
random_selection.to_csv(output_file_path, index=False)

print(f"Output file saved as: {output_file_path}")
```
